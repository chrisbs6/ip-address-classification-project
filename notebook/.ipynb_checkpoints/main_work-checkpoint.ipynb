## 2. Extracting Per-Address Metrics

```python
def compute_metrics_rtt(group, time_col, rtt_col, probe_interval=660):
    """
    For each address (group), compute:
      - availability: mean of is_up (0 or 1)
      - median_RTT_us: median RTT among probes with RTT > 0.
    If no positive RTT is recorded, returns NaN.
    """
    group = group.sort_values(time_col)
    
    availability = group['is_up'].mean()
    
    rtts = group[rtt_col].values
    positive_rtts = rtts[rtts > 0]
    median_RTT = np.median(positive_rtts) if positive_rtts.size > 0 else np.nan
    
    return pd.Series({
        'availability': availability,
        'median_RTT_us': median_RTT
    })
```

## 4. Block-Level Clustering (BlockSizeId)

```python
def ip_to_int(ip_str):
    return int(ipaddress.IPv4Address(ip_str))

def block_size_id(df, current_prefix, epsilon=2.0):
    df = df.copy()
    df['ip_int'] = df['probe_addr'].apply(ip_to_int)
    df = df.sort_values('ip_int')
    
    overall_var = df[['availability', 'median_RTT_us']].var().sum()
    
    best_prefix = current_prefix
    best_var_sum = overall_var
    
    for p in range(current_prefix + 1, 33):
        sub_block_size = 2 ** (32 - p)
        base = df['ip_int'].min()
        df['sub_block'] = ((df['ip_int'] - base) // sub_block_size)
        
        var_sum = 0
        for _, group in df.groupby('sub_block'):
            var_sum += group[['availability', 'median_RTT_us']].var().sum()
        
        if best_var_sum - var_sum < epsilon:
            break
        else:
            best_var_sum = var_sum
            best_prefix = p

    if best_prefix == current_prefix:
        return [df]
    else:
        sub_block_size = 2 ** (32 - best_prefix)
        base = df['ip_int'].min()
        df['sub_block'] = ((df['ip_int'] - base) // sub_block_size)
        
        consistent_blocks = []
        for _, group in df.groupby('sub_block'):
            consistent_blocks.extend(block_size_id(group.copy(), best_prefix + 1, epsilon))
        return consistent_blocks
```

```python
def process_all_blocks(metrics_df, ip_col, block_col='block_24', initial_prefix=24, epsilon=2.0):
    consistent_blocks_all = []
    unique_blocks = metrics_df[block_col].unique()
    
    for block_id in tqdm(unique_blocks, desc="Processing /24 blocks"):
        group = metrics_df[metrics_df[block_col] == block_id]
        sub_df = group.copy()
        sub_df = sub_df.rename(columns={ip_col: 'probe_addr'})
        consistent_blocks = block_size_id(sub_df, current_prefix=initial_prefix, epsilon=epsilon)
        for block in consistent_blocks:
            block_size = block.shape[0]
            block['final_block_size'] = block_size
            block['original_block'] = block_id
            consistent_blocks_all.append(block)
            
    return pd.concat(consistent_blocks_all, ignore_index=True)
```

# Clustering

The interpretation in our classifier is a heuristic based on the assumption that, within the ambiguous range of availability (between 0.2 and 0.95), blocks with higher RTT tend to be associated with slower, older, or dial‐up connections that remain up more consistently—thus they are labeled "sometimes‑active"—while those with lower RTT are often fast broadband connections that may only be active briefly, hence "intermittent."

In other words, if an ambiguous block’s median RTT is higher than the overall median among active blocks, we assume it’s slower (and perhaps more continuously connected, like dial‑up or legacy connections that are always on despite their slower speeds). Conversely, a block with a lower median RTT might indicate a broadband connection that, while fast when active, is not up as continuously.

Keep in mind that this mapping is a data‐driven approximation and may need to be tuned based on your dataset. The idea is to use RTT as a surrogate for the more detailed metrics (volatility and median up time) from the Cai paper, which aren’t available in your sparse data.

```python
def classify_blocks_ml(blocks_df):
    """
    Classify blocks using an ML-based approach with two features:
      - availability (mean, 0 to 1)
      - median_RTT_us (median RTT in microseconds computed over responsive probes)
      
    Blocks with availability < 0.2 are labeled as 'inactive'.
    
    Active blocks (availability >= 0.2 and non-null median_RTT_us) 
    are clustered into 4 clusters. The clusters are then interpreted as follows:
      - The cluster with the highest availability is 'always-active'
      - The cluster with the lowest availability is 'underutilized'
      - For the two remaining clusters, compare their median_RTT_us to the overall median:
           * If higher, label as 'sometimes-active'
           * Otherwise, label as 'intermittent'
    
    Returns:
      all_classified: DataFrame with an added 'category' column.
      centers: DataFrame of cluster centers (in original feature space).
    """
    import pandas as pd
    from sklearn.preprocessing import StandardScaler
    from sklearn.cluster import KMeans
    
    df = blocks_df.copy()
    
    # Separate blocks into inactive and active.
    inactive = df[df['availability'] < 0.2].copy()
    inactive['category'] = 'inactive'
    
    # For active blocks, require that median_RTT_us is not NaN.
    active = df[(df['availability'] >= 0.2) & (df['median_RTT_us'].notna())].copy()
    
    # Use features: availability and median_RTT_us.
    X = active[['availability', 'median_RTT_us']].values
    
    # Scale the features.
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Cluster active blocks into 4 clusters.
    kmeans = KMeans(n_clusters=4, random_state=42)
    active['cluster_id'] = kmeans.fit_predict(X_scaled)
    
    # Inverse-transform the cluster centers for interpretation.
    centers = pd.DataFrame(
        scaler.inverse_transform(kmeans.cluster_centers_),
        columns=['availability', 'median_RTT_us']
    )
    
    # Map clusters to categories:
    # Sort clusters by availability.
    sorted_idx = centers.sort_values(by='availability', ascending=False).index
    mapping = {}
    mapping[sorted_idx[0]] = 'always-active'
    mapping[sorted_idx[-1]] = 'underutilized'
    
    # For the two middle clusters, use median_RTT_us relative to overall median.
    overall_median_rtt = active['median_RTT_us'].median()
    for idx in sorted_idx[1:-1]:
        if centers.loc[idx, 'median_RTT_us'] >= overall_median_rtt:
            mapping[idx] = 'sometimes-active'
        else:
            mapping[idx] = 'intermittent'
    
    active['category'] = active['cluster_id'].map(mapping)
    
    # Combine active and inactive blocks.
    all_classified = pd.concat([active, inactive])
    
    return all_classified, centers
```

```python
block_metrics_2013['probe_addr_2013'] = block_metrics_2013['block_24']

# Call the classifier function.
classified_blocks_2013, centers_2013 = classify_blocks_ml(block_metrics_2013)

print("Classified 2013 Blocks:")
print(classified_blocks_2013[['probe_addr_2013', 'availability', 'median_RTT_us', 'block_24', 'category']].head())
```

```python
def cai_classifier_dynamic(block, overall_median_rtt):
    """
    Classify a block based on availability and median_RTT_us using Cai-inspired thresholds.
    
    Blocks with availability < 0.2 are unclassifiable.
    Blocks with availability ≥ 0.95 are always-active.
    Blocks with availability between 0.2 and 0.95 are classified using a dynamic RTT threshold.
    
    Parameters:
      block: a row (Series) representing a block's metrics.
      overall_median_rtt: the median of median_RTT_us computed over blocks with availability in [0.2, 0.95].
      
    Returns:
      A string label.
    """
    if block['availability'] < 0.2:
        return 'unclassifiable'
    elif block['availability'] >= 0.95:
        return 'always-active'
    else:
        # For blocks with availability between 0.2 and 0.95,
        # use the overall median RTT as the threshold.
        if block['median_RTT_us'] >= overall_median_rtt:
            return 'sometimes-active'
        else:
            return 'intermittent'
```

```python
ambiguous_2013 = block_metrics_2013[
    (block_metrics_2013['availability'] >= 0.2) & (block_metrics_2013['availability'] < 0.95)
]
# Compute overall median RTT among ambiguous blocks.
overall_median_rtt_2013 = ambiguous_2013['median_RTT_us'].median()

# Apply the classifier row-by-row.
block_metrics_2013['cai_category'] = block_metrics_2013.apply(
    lambda row: cai_classifier_dynamic(row, overall_median_rtt_2013), axis=1
)

print("Cai threshold-based classifier (2013):")
print(block_metrics_2013['cai_category'].value_counts(normalize=True))
```

```python
# For 2023:
ambiguous_2023 = block_metrics_2023[
    (block_metrics_2023['availability'] >= 0.2) & (block_metrics_2023['availability'] < 0.95)
]
overall_median_rtt_2023 = ambiguous_2023['median_RTT_us'].median()

block_metrics_2023['cai_category'] = block_metrics_2023.apply(
    lambda row: cai_classifier_dynamic(row, overall_median_rtt_2023), axis=1
)

print("Cai threshold-based classifier (2023):")
print(block_metrics_2023['cai_category'].value_counts(normalize=True))
```
